```python
with open("followers.txt","r",encoding="utf-8") as file:
    
    followers = (file.read()).split("\n")
    
with open("followings.txt","r",encoding="utf-8") as file2:
    followings = (file2.read()).split("\n")


thoseWhoDontFollow = []
    
for following in followings:
    if following not in followers:
        if following not in thoseWhoDontFollow:
            thoseWhoDontFollow.append(following)
        

print(len(thoseWhoDontFollow))

for person in thoseWhoDontFollow:
    print(person)
```
